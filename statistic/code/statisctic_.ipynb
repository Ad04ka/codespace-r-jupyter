```python
#7.1. For file data House.txt construct an equation of linear paired and multiple linear regression linking
#the living area of the house, its age and estimated value.
#• Investigate the statistical significance of the constructed equation.

m1<-lm(Price~SqFt, data = House)
m1
#1.Model
#Price= 51.92+16.63*SqFt

#2.Interpretation
# b1=16.63:  If area increases by 1 SqFt then the Price will increase by 16.63 $thousand in average
# b0=51.92: additional coast 

 #3.Quality of the model

#3.1 Statistical significant of coefficients (Student  Test)
#Ho: b1=0 (b1 is insignificant)
#Ha: b1 NOT=0 (b1 is significant)

summary(m1)

#Coefficients:
#                Estimate         Std. Error      t value    Pr(>|t|)    
#(Intercept)     b0^=51.915      SE(bo)=5.563      9.333     3.98e-07 < 0,05 We reject Ho, b0 is significant
#  SqFt          b1^=16.633      SE(b1)=3.316      5.016     0.000236 < 0,05 We reject Ho, b1 is significant

# Signif. codes:  0 ‘***’ 0.001 ‘**’ 0.01 ‘*’ 0.05 ‘.’ 0.1 ‘ ’ 1

51.915/5.563
#9.332195
16.633/3.316 
#5.015983


#3.2 Satistical significant of the model (Fisher Test)

#Coefficient of determination
#Multiple R-squared:  0.6593. variation of y by 66% is explained  by variation of x

#H0: R^2=0 model is insignificant
#Ha: R^2 NOT =0 model is significant

#F-statistic: 25.16 on 1 and 13 DF,  p-value: 0.0002362 < 0,05 H0 is rejected, model is significant
0.6593/(1-0.6593)*(15-2)
#25.15674


#3.3 MAPE 
e<-m1$residuals
e

Price_m1<-m1$fitted.values
Price_m1
mean(abs(e/House$Price))*100
#2.832387 <10% we we construct a  good approximation

install.packages("MLmetrics")
library(MLmetrics)
MAPE(m1$fitted.values, House$Price)*100


#2.832387

plot(House$SqFt, House$Price)
abline(m1)

#3.4 Regression assumptions

#3.4.1. E(erros)=0

#H0: E(e)=0
#Ha: E(e) Not=0

t.test(m1$residuals, alternative = "t", conf.level = 0.95)
#p-value = 1 there is no reason to reject Ho, the first assuption is fulfild

#3.4.2. Homoskedastisity (Goldfeld–Quandt)
install.packages("lmtest")
library(lmtest)

#Ho: r(e x)=0 (variances are constant) or  Homoskedastisity
#Ho: r(e x)NOT =0 (variances are NOT constant) or  Heteroskedastisity
gqtest(m1, alternative = "t")
#p-value = 0.6157> 0.05, there is no reason to reject Ho, so Homoskedastisity

#Broysh–Pagan 

bptest(m1,studentize = FALSE)
#p-value = 0.6535 there is no reason to reject Ho, so Homoskedastisity

#3.4.3 NO autocorrelation
#the  first order of autocorrelation Darbin–Watson

# H0: r(e(i), e(i+1))=0  there is NO  autocorrelation  of the 1 order
# H0: r(e(i), e(i+1)) NOt =0  there is autocorrelation  of the 1 order

dwtest(m1)
#p-value = 0.619>0,05 there is NO  autocorrelation  of the 1 order

#Broich–Godfrey for any odrder of autocorrelation 
#H0: r1=r1=r3=0  there is NO  autocorrelation  of the 1, 2, 3  orders
#Ha:   at least one autocorrelation coefficients Not 0

bgtest(m1, order = 3)
#p-value = 0.9107>0,05  there is NO  autocorrelation  of the 1, 2, 3  orders

#Homework: construct simple linear regression for any data from suppermarket sales
```
